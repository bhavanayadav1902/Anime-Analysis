## DATA CLEANING:

#### Imputing the null values of the finishYr column by finding the mode: 

```python
count1=0
count2=50
for i in range(13):
    list_mode= data.finishYr[count1:count2].mode()[0]
    data.finishYr[count1:count2]= data.finishYr[count1:count2].fillna(list_mode)
    count1+=50
    count2+=50
```

#### Imputing the null values of the sznOfRelease column by finding the mode since it is categorical data:

```python
count1=0
count2=50
for i in range(13):
    list_mode = data.sznOfRelease[count1:count2].mode()[0]
    data.sznOfRelease[count1:count2]= data.sznOfRelease[count1:count2].fillna(list_mode)
    count1+=50
    count2+=50
```

##### Imputing the null values of the watched column by finding the mean:

```python
watched_mean=data.watched.mean()
data.watched=data.watched.fillna(watched_mean)
```

##### Imputing the null values of the watching column by finding the average of the above and before values i.e interpolation:

```python
data.watching=data.watching.interpolate()
```

#####  Imputing the null values of the votes column by finding the median since this column has some outliers:

```python
plt.boxplot(data['votes'], vert = False)
plt.xlabel('votes')
```

```python
votes_median=data.votes.median()
data.votes=data.votes.fillna(votes_median)
```

##### Imputing the null values of the dropped column by finding the average of the above and before values i.e interpolation:

```python
data.dropped=data.dropped.interpolate()
```

##### Imputing the null values of the studios column by finding the mode since it is categorical data:

```python
count1=0
count2=50
for i in range(13):
    list_mode= data.studios[count1:count2].mode()[0]
    data.studios[count1:count2]= data.studios[count1:count2].fillna(list_mode)
    count1+=50
    count2+=50
```

##### the column "contentWarn" is dropped because it has large number of missing values:

```python
data.drop(["contentWarn"], axis = 1, inplace = True )
data
```

##### Imputing the null values of the mediaType column by finding the mode since it is categorical data :

```python
mediaType_mode=data.mediaType.mode()[0]
data.mediaType=data.mediaType.fillna(mediaType_mode)
```

##### The column "duration" is dropped because it has large number of missing values:

```python
data.drop(["duration"], axis = 1, inplace = True )
```

##### imputing the null value present in the "eps" column by inperpolate method since we have single null value:

```python
data.eps=data.eps.interpolate()
```

## HYPOTHESIS TESTING:

### Test if sample of votes is from the population of votes

```python
votes=data['votes']
mean_votes=votes.mean()
sigma_votes=votes.std(ddof=0)
print("mean: ", mean_votes, ", sigma:", sigma_votes)
```

```python
sample=votes.sample(n=300)
x_bar=sample.mean()
print("sample_mean: ",x_bar)
```

###### H0: The sample is from the anime population, x_bar = μ(mean)
###### HA: The sample is not from the anime population, x_bar != (not equal) μ(mean)
###### If the p-value is greater than alpha then we say H0 is plausible , otherwise we reject the null and accept the alternative hypothesis.
###### alpha level of α = 0.05

```python
# alpha = 0.05
z_critical = 1.96 # alpha level of 0.05 and two-tailed test
x_bar = 6234.243333333333
N = 300
SE = sigma_votes/np.sqrt(N)
z_stat = (x_bar - mean_votes)/SE
print(z_stat)
```

###### p-value = 0.03*2 = 0.06(since two tailed test) (from z-table)
###### Since p-value > alpha it is plausible for H0 to be true and therefore we  reject the althernative
###### Statistically, we say the sample mean is no different than the population mean and thus the sample is drawn from the population
